```python
import os
import json
import pathlib
```

```python
CAPTIONS_FILE = "/scratch/govind/modality_translation/llava_caption.json"

with open(CAPTIONS_FILE) as f:
    captions = json.load(f)
```

```python
captions
```

```python
captions_fix = {}

for path, caption in captions.items():
    caption_fix = caption.replace("sks ", "")
    path_fix = path.replace("images_rgb_train", "images_thermal_train")
    captions_fix[path_fix] = caption_fix
```

```python
captions_fix
```

```python
for path, caption in captions_fix.items():
    image_path = pathlib.Path(path)
    captions_path = image_path.parent / (image_path.stem + ".txt")
    print(image_path, captions_path)
```
