```python
# load the tiny stories dataset
import os
import pathlib
import pandas as pd
```

```python
df = pd.read_csv(pathlib.Path('.') / 'validation.csv')
```

```python
texts = df['text'].tolist()
```

```python
print(texts[4])
```

```python
# print every distinct word in the dataset
words = set()
for text in texts:
    # remove all punctuation
    text = text.replace('.', '').replace(',', '').replace('!', '').replace('?', '').replace(';', '').replace(':', '').replace('"', '')
    text = text.lower()
    for word in text.split():
        words.add(word)
```

```python
for word in sorted(words):
    if word[-3:] == 'ing':
        print(word)
```
