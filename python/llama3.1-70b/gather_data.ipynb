```python
from gradio_client import Client
import pandas as pd
import numpy as np
from sklearn import datasets, decomposition
import matplotlib.pyplot as plt
```

```python
client = Client("open-llm-leaderboard/open_llm_leaderboard")
result = client.predict(
		api_name="/get_latest_data_leaderboard"
)
```

```python

df = pd.DataFrame(result["data"], columns=result["headers"])
```

```python
df
```

```python
df.columns
```

```python
df[df["Maintainer's Highlight"] == True]
```

```python
df_raw_scores = df[['IFEval Raw', 'BBH Raw', 'MATH Lvl 5 Raw', 'GPQA Raw', 'MUSR Raw']]
df_raw_scores
```

```python
df_raw_scores_corr = df_raw_scores.corr()
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(df_raw_scores_corr, cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ax.set_xticks(np.arange(len(df_raw_scores_corr.columns)))
ax.set_xticklabels(df_raw_scores_corr.columns)
ax.set_yticks(np.arange(len(df_raw_scores_corr.columns)))
ax.set_yticklabels(df_raw_scores_corr.columns)
```

```python
pca = decomposition.PCA(n_components=3)
pca.fit(df_raw_scores)
```

```python
pca.components_
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(pca.components_, cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ax.set_title("PCA Components")
ax.set_xticks([0, 1, 2, 3, 4])
ax.set_xticklabels(df_raw_scores.columns)
ax.set_yticks([0, 1, 2])
ax.set_yticklabels(["PC1", "PC2", "PC3"])
```

```python
# Explained variance
plt.bar([f"PCA{n+1}" for n in range(pca.n_components_)], pca.explained_variance_ratio_)
plt.title("Explained Variance")
plt.ylabel("Explained Variance Ratio")
```

```python
df_raw_scores_transformed = pca.transform(df_raw_scores)
df_raw_scores_transformed
```
